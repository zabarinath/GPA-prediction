```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv(r"C:\Users\007sa\Downloads\1.03.+Dummies.csv")
```

```python
data = raw_data.copy()
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data.describe()
```

## REGRESSION


```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x =sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
